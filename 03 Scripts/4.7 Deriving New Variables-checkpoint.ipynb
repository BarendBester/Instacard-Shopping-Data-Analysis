2. Create your new function

```python
# Create a new function to catogorize prices 
def price_lab(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid_range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# add new row and use our new function
df['price_range'] = df.apply(price_lab, axis=1)
```

```python
#check the outcome
df.head()
```

```python
#check number of price range values
df['price_range'].value_counts(dropna = False)
```

3. if statements with the loc function

```python
#set low range product
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. 

```python
# step 1 recatogorize orders into different periods of the day
# count the orders per hour
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
# 10-16 most busy
# 0-6 least busy
# 7-9 and 17-23 average
```

```python
#split these into 3 periods using loc function
#1. least orders
ords_prods_merged.loc[ords_prods_merged['order_hour_of_day'] <= 6, 'busiest_period_of_day'] = 'Fewest orders'
```

```python
#2. most orders
ords_prods_merged.loc[(ords_prods_merged['order_hour_of_day'] >= 10) & (ords_prods_merged['order_hour_of_day'] <= 16) , 'busiest_period_of_day'] = 'Most orders'
```

```python
#3 average orders
ords_prods_merged.loc[(ords_prods_merged['order_hour_of_day'] > 6) & (ords_prods_merged['order_hour_of_day'] < 10) & (ords_prods_merged['order_hour_of_day'] >= 17) , 'busiest_period_of_day'] = 'Average orders'
```
