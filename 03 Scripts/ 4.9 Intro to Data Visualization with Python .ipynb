```python
#import dataset
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_agg_grouped.pkl'))
```

1. **Creating Bar Charts**

```python
#Plot a bar graph
bar_graph_basic = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
#download figure
bar_graph_basic.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

2. **Create Histograms**

```python
histogram_plot_basic = ords_prods_merge['prices'].plot.hist(bins = 25)
```

3. **Scatter plots**

```python
# Use scatterplot to identify outliers
scatter_plot_basic = sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge)
```
