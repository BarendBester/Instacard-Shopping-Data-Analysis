4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
#Split the data into groups based on the “user_id” column.
#Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
#Create a new column, “max_order,” into which you’ll place the results of your aggregation.
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(10)
```

```python
#use loc function to assign loyalty
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#use loc function to assign average
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#use loc function to assign new customer
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#use valuecounts function to check loyalty of customers
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

6. The team now wants to target different types of spenders in their marketing campaigns.

```python
#1. create spending flag based on avg price
#Split the data into groups based on the “user_id” column.
#Apply the transform() function on the “prices” column to generate the avgerage price for each user.
#Create a new column,into which you’ll place the results of your aggregation.
ords_prods_merge['avg_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(5)
```

```python
#use loc function to assign spending flag
ords_prods_merge.loc[ords_prods_merge['avg_spend'] < 10, 'spender_type'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_spend'] >= 10, 'spender_type'] = 'High spender'
```

```python
ords_prods_merge.head(20)
```

```python
ords_prods_merge['spender_type'].value_counts(dropna = False)
```

7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers.

```python
#1. create spending flag based on median orders
#Split the data into groups based on the “user_id” column.
#Apply the transform() function on the “day since prior order” column to generate the median orders for each user.
#Create a new column, into which you’ll place the results of your aggregation.
ords_prods_merge['median_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check your flag
ords_prods_merge.head(20)
```

```python
#use loc function to assign frequency of orders flag
ords_prods_merge.loc[ords_prods_merge['median_orders'] > 20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_orders'] > 10) & (ords_prods_merge['median_orders'] <= 20) , 'customer_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_orders'] <= 10, 'customer_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge.head(20)
```

```python
#check frequency of customer frequency
ords_prods_merge['customer_frequency'].value_counts(dropna = False)
```
